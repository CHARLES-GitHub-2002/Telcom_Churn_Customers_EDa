### Insignts of Univariate Analysis

2. Convertion of churn variable into a binary numeric variable yes=1,no=0

```python
# Clean the column for consistent formatting
New_df['Churn'] = New_df['Churn'].astype(str).str.strip().str.lower()
# Replace 'yes' with 1 and 'no' with 0
new_df = New_df.replace({'Churn': {'yes': 1, 'no': 0}})
# Explicitly downcast (optional if you rely on old behavior)
new_df = new_df.infer_objects(copy=False)
# Check the first few rows
new_df.head()
```

3. Converting Categolical Variable In to dummy variables
- *Dummy variables (also called indicator variables) are binary variables used to represent categorical data in a form that can be easily understood by machine learning algorithms or statistical models.*

```python
# Create dummy variables
df_dummies = pd.get_dummies(new_df)
# Convert all columns in df_dummies to integers (1 and 0)
df_dummies = df_dummies.astype(int)
df_dummies.head()
```

4. Relationship Between Monthly Charges and Total Charges

```python
sns.scatterplot(data=df_dummies, x='MonthlyCharges', y='TotalCharges')
# Add title and labels
plt.title("Monthly Charges vs. Total Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Total Charges")
# Show plot
plt.show()
```

Total charges increse as monthly charges increse - as expected 

5. Build a correration of all predictors with Churn

```python
# Plot
# Calculate the correlation matrix and extract correlations with 'Churn'
correlations = df_dummies.corr()['Churn'].sort_values(ascending=False)
# Plot the correlations
plt.figure(figsize=(20, 8))
correlations.plot(kind='bar', color='blue', edgecolor='black')# Add title and labels
plt.title("Feature Correlation with Churn", fontsize=16)
plt.ylabel("Correlation Coefficient", fontsize=12)
plt.xlabel("Features", fontsize=20)
# Show the plot
plt.tight_layout()
plt.show()
```

**Derived Insight:**

- HIGH Churn seen in case of Month to month contracts, No online security, No Tech support, First year of subscription and Fibre Optics Internet

- LOW Churn is seens in case of Long term contracts, Subscriptions without internet service and The customers engaged for 5+ years

- Factors like Gender, Availability of PhoneService and # of multiple lines have alomost NO impact on Churn

This is also evident from the Heatmap below

```python
# Plot the heatmap

plt.figure(figsize=(12,12))
sns.heatmap(df_dummies.corr(), cmap="Paired")
# Add title for the heatmap
plt.title("Correlation Matrix Heatmap", fontsize=16)
# Show the plot
plt.tight_layout()
plt.show()
```

# Bivariate Analysis

```python
new_df1_target0=new_df.loc[new_df["Churn"]==0]
new_df1_target1=new_df.loc[new_df["Churn"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = new_df, x= col, order=new_df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target1,col='PaymentMethod',title='Distribution of PaymentMethod for Churned Customers',hue='gender')
```
